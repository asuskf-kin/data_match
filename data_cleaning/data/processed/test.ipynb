```python
# 1. Tu regex original
regex = (
    r"\b(?:"
    r"benavides\s+pharmacies|"
    r"bimbo.*|"
    r"café\s+la\s+flor\s+de\s+córdoba|"
    r"chilaquiles\s+la\s+chilaquila|"
    r"croissants\s+alfredo|"
    r"danny\s+yo|"
    r"dolce\s+natura|"
    r"el\s+pechugon|"
    r"el\s+pollo\s+bronco|"
    r"el\s+pollo\s+giro.*|"
    r"farmacia\s+dermatologica|"
    r"farmacia\s+homeopatica|"
    r"farmacia\s+similares|"
    r"farmacia\s+de\s+similares|"
    r"farmacia\s+dg|"
    r"helados\s+dolphy|"
    r"dolphy|"
    r"ok\s+pasteleria|"
    r"pasteleria\s+ok|"
    r"pastelerias\s+marisa|"
    r"pizza\s+los\s+arcos|"
    r"pizzas\s+los\s+arcos|"
    r"pasteleria\s+la\s+michoacana|"
    r"pasteleria\s+michoacana|"
    r"pollo\s+pechugon|"
    r"pollo\s+bronco|"
    r"pollo\s+pepe|"
    r"rosticeria\s+el\s+pechugon|"
    r"rosticerias\s+el\s+pechugon|"
    r"tia\s+lola\s+pasteleria|"
    r"nescafé.*"
    r")\b"
)

# 2. Normalizar a minúsculas
df5["name_2"] = df5["name_normalized"].str.lower()

# 3. Detectar si hace match con el regex (lo guardamos en una variable temporal, no sobrescribimos identified_as_chain)
match_regex = df5["name_2"].str.contains(regex, regex=True, na=False)

# Extraemos el texto para que puedas validar qué encontró
df5["matched_text"] = df5["name_2"].str.extract(f"({regex})", expand=False)

# 4. FLUJO DE VALIDACIÓN: Creamos la condición estricta (Match Regex Y identified_as_chain == True)
condicion_eliminar = (match_regex == True) & (df5["identified_as_chain"] == True)

print("--- REGISTROS QUE SE VAN A ELIMINAR ---")
validaciones = df5[condicion_eliminar][["name_2", "matched_text", "identified_as_chain"]]
print(validaciones.head(20)) 
print(f"\nTotal de registros a eliminar: {len(validaciones)}")

# 5. ELIMINAR los que cumplen AMBAS condiciones
# Usamos el símbolo ~ para decirle a pandas: "Quédate con todo lo que NO cumpla la condicion_eliminar"
df5_final = df5[~condicion_eliminar].copy()

# 6. Limpieza final: Eliminar el '=' al inicio de la columna 'name' y borrar columnas auxiliares
df5_final["name"] = df5_final["name"].str.replace(r"^=", "", regex=True).str.strip()
df5_final = df5_final.drop(columns=["matched_text", "name_2"])

print(f"Total de registros conservados en df5_final: {len(df5_final)}")
```
